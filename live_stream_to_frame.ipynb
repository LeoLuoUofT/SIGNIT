```python
from pyspark.sql import SparkSession
import subprocess

def get_stream_url(url):
    try:
        result = subprocess.run(['youtube-dl', '-g', url], stdout = subprocess.PIPE)
        return result.stdout.decode('utf-8').strip()
    except Exception as e:
        print(f"Error fetching stream URL: {e}")
        return None

def pull_frames(stream_url):
    try:
        subprocess.run(['ffmpeg', '-i', stream_url, '-vf', 'fps=10', '-t', '300', 'output2/out1-%4d.png'])
    except Exception as e:
        print(f"Error downloading frames: {e}")

def process_video(url):
    stream_url = get_stream_url(url)
    if stream_url:
        pull_frames(stream_url)
    return f"Processed {url}"

if __name__ == "__main__":

 spark = SparkSession.builder.appName("Stream Downloader").getOrCreate()

 URL_List = ["https://www.youtube.com/live/gCNeDWCI0vo?si=AAxtZcpFBL26CbEG","https://www.twitch.tv/clix"]

 URL_List_rdd = spark.sparkContext.parallelize(URL_List)

 results = URL_List_rdd.map(process_video).collect()

 print(results)

 spark.stop()
```
